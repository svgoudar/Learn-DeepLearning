
## Know about Images

### Grayscale Images

* **Definition:** Images with only **one color channel**. Also called black-and-white images.
* **Representation:** Height × Width × 1 (e.g., a 6×6 grayscale image is 6×6×1).
* **Pixel values:** Each pixel has an intensity value ranging from 0 (black) to 255 (white).
* **Example:** Classic black-and-white photos or early movies.

**Key point:** Since there’s only one channel, each pixel stores a single intensity value.

---

### **2. RGB (Colored) Images**

* **Definition:** Images composed of **three color channels**:

  1. **R (Red)**
  2. **G (Green)**
  3. **B (Blue)**
* **Representation:** Height × Width × 3 (e.g., a 4×4 RGB image is 4×4×3).
* **Pixel values:** Each pixel in each channel ranges from 0 to 255.
* **How it works:** Combining the three channels produces the full-color image.

**Example:** A colored photo of a parrot or landscape:

* **Red channel:** Highlights red intensity.
* **Green channel:** Highlights green intensity.
* **Blue channel:** Highlights blue intensity.
* **Combination:** Full-color visualization.

---

### **3. Pixels and Image Dimensions**

* Images are made up of tiny units called **pixels**.
* **Height × Width** = number of pixels.
* **Channels** = depth of the image (1 for grayscale, 3 for RGB).
* **Notation:**

  * Grayscale → H×W×1
  * RGB → H×W×3

**Example:**

* 6×6 grayscale → 6×6×1
* 4×4 RGB → 4×4×3

---

### **4. Relevance for CNNs**

* CNNs process images **numerically**.
* Each pixel’s value acts as input data for the network.
* Understanding **channels** is important because convolution operations are applied **per channel**, and features are extracted from either single-channel (grayscale) or multi-channel (RGB) images.

---

**Summary**

1. **Grayscale:** One channel, pixel values 0–255.
2. **RGB:** Three channels (R, G, B), each pixel in each channel 0–255.
3. Image shape: Height × Width × Channels.
4. CNNs use these pixel values to **extract features** via convolution operations.
