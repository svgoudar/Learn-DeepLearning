```
1 2
4 3  → min = 1
```

---

### 4. Average Pooling

* Take **average** of values in each region.
* Used when you want a smooth representation instead of strongest or weakest activations.

Example:

```
1 2
4 3  → avg = 2.5
```

---

### 5. Location Invariance

* Suppose a filter detects a feature (e.g., cat face, wheel).
* Even if the object shifts slightly in the image, max pooling captures the important activation.
* Pooling ignores exact position and keeps **presence** of the feature.

---

### 6. Typical stack

A CNN block often looks like:

```
Input → Convolution → ReLU → Pooling → (repeat) → Fully Connected Layer
```

You repeat convolution + pooling multiple times. At the end, flatten and pass to dense layers.

---
